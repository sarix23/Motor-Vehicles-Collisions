```python
import pandas as pd

weather = pd.read_csv('Weather.csv') #2119920
# Rounding the temperature column
weather['temperature_2m'] = weather['temperature_2m'].round().astype(int)

# Rounding the humidity column
weather['relative_humidity_2m'] = (weather['relative_humidity_2m'] // 10) / 10

# Splitting the date column into date and time
weather[['date', 'time']] = weather['date'].str.split(' ', expand=True)

# Reorganizing the columns
weather = weather[['date', 'time', 'zip_code', 'temperature_2m', 'relative_humidity_2m', 'rain', 'snowfall', 'cloud_cover']]

# Define functions for value replacement
def replace_rain(value):
    value = float(value)
    if 0.0 == value:
        return "No Rain"
    if 0.0 < value:
        return "Rain"

def replace_snowfall(value):
    value = float(value)
    if 0.0 == value:
        return "No Snow"
    if 0.0 < value:
        return "Snow"

def replace_cloud_cover(value):
    value = int(value)
    if 0 <= value < 10:
        return "No Cloud"
    elif 10 <= value:
        return "Cloud"

# Apply replacement functions to respective fields
weather['rain'] = weather['rain'].apply(replace_rain)
weather['snowfall'] = weather['snowfall'].apply(replace_snowfall)
weather['cloud_cover'] = weather['cloud_cover'].apply(replace_cloud_cover)

# Save the modified DataFrame to a new CSV file
weather.to_csv('weather_cleaned.csv', index=False) #2119920
```
