```python
import pandas as pd
import numpy as np

vehicles = df = pd.read_csv('Vehicles.csv')

print(len(vehicles))

# Save the collision_id of rows to be deleted
collision_id_deleted = set(vehicles.loc[vehicles['pre_crash'].isin([np.nan]), 'collision_id'].tolist())

# Delete rows with collision_id present in the collision_id_deleted list
vehicles = vehicles[~vehicles['collision_id'].isin(collision_id_deleted)]

# Transform crash_date from '2018-01-01T00:00:00.000' to '2018-01-01'
vehicles['crash_date'] = pd.to_datetime(vehicles['crash_date']).dt.strftime('%Y-%m-%d')

# Transform crash_time from '3:36' to '03:00:00' or '16:53' to '16:00:00'
vehicles['crash_time'] = pd.to_datetime(vehicles['crash_time'], format='%H:%M').dt.strftime('%H:00:00')

# Print the first few records of the transformed DataFrame
print(len(vehicles))
print(vehicles.head())

# Print the collision_id of deleted rows
print("Collision IDs of deleted rows:", collision_id_deleted)

# Save the modified DataFrame to a new CSV file
vehicles.to_csv('vehicles_cleaned.csv', index=False)

```
